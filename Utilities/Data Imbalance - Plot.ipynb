```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('/kaggle/input/fraud-transaction-detection/Final Transactions.csv')
```

```python
count1 = 0 
count0 = 0
for i in df['TX_FRAUD'].values:
    if i == 1:
        count1 += 1
    else:
        count0 += 1
        
count1 = (count1/len(df['TX_FRAUD']))*100
count0 = (count0/len(df['TX_FRAUD']))*100

x = ['Fraudulent Transactions','Genuine Transactions']
y = [count1, count0]


explode = (0.1, 0)  # only "explode" the 1st slice

fig1, ax1 = plt.subplots()
ax1.pie(y,
        explode=explode,
        labels=x,
        autopct='%1.1f%%',
        shadow=True,
        startangle=110)
ax1.axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle.
plt.title('Pie Chart - Data Imbalance',fontsize=25)

plt.savefig('PieChart-DataImbalance.png')
plt.show()
```
